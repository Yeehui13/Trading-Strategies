```python
lags = 2 #history for 2 days
```

```python
def create_lags(data):
    global cols
    cols = []
    for lag in range(1, lags + 1):
      col = 'lag_{}'.format(lag)
      data[col] = data['return'].shift(lag)
      cols.append(col)
```

Frequency Approach

```python
def create_bins(data, bins=[0]):
    global cols_bin
    cols_bin = []
    for col in cols:
        col_bin = col + '_bin'
        data[col_bin] = np.digitize(data[col], bins=bins)
        cols_bin.append(col_bin)
```

DNN with Keras & TensorFlow Backend

```python
import tensorflow as tf
from keras.layers import Dense
from keras.models import Sequential
```

```python
def create_model():
  np.random.seed(100)
  tf.random.set_seed(1000)
  model = Sequential()
  model.add(Dense(16, activation="relu", input_dim=lags))
  model.add(Dense(16, activation="relu"))
  model.add(Dense(1, activation="sigmoid"))
  model.compile(loss="binary_crossentropy", optimizer="rmsprop",
                metrics=["accuracy"])
  return model
```

```python
data_ = (data - data.mean()) / data.std()
data['direction_'] = np.where(data['direction'] == 1, 1, 0)
```

```python
model = create_model()
```

```python
%%time
model.fit(data_[cols], data['direction'],
          epochs=50, verbose=False)
```

```python
model.evaluate(data_[cols], data['direction_'])
```

```python
pred = np.where(model.predict(data_[cols]) > 0.5, 1, 0)
```

```python
data['pos_dnn_ke'] = np.where(pred > 0, 1, -1)
```

```python
data['strat_dnn_ke'] = data['pos_dnn_ke'] * data['return']
```
